# Modelling CO2

<br>

### Imports

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, RationalQuadratic, ExpSineSquared
```

<br>

### Helper Functions

```python
def hide_spines(ax, positions=['top', 'right']):
    """
    Pass a matplotlib axis and list of positions with spines to be removed
    
    args:
        ax:          Matplotlib axis object
        positions:   Python list e.g. ['top', 'bottom']
    """
    assert isinstance(positions, list), 'Position must be passed as a list '
    
    for position in positions:
        ax.spines[position].set_visible(False)
        
def drop_ticks(ax):
    ax.set_xticks([])
    ax.set_yticks([])
    
def rgb_2_plt_tuple(rgb_tuple):
    plt_tuple = tuple([x/255 for x in rgb_tuple])
    return plt_tuple
```

```python
df_co2 = pd.read_csv('data/CO2/cleaned_mauna_loa.csv', parse_dates=['datetime'])

df_co2.head()
```

```python
## Defining the target variable
y = df_co2['CO2_ppm'].values

## Formatting the feature variable
s_dt = df_co2['datetime']
s_year_frac = s_dt.dt.year + s_dt.dt.dayofyear/365

X = s_year_frac.values.reshape(-1, 1)
X
```

```python
# long term smooth rising trend
k1 = 50.0**2 * RBF(length_scale=50.0) 

# seasonal component
k2 = (2.0**2 
      * RBF(length_scale=100.0) 
      * ExpSineSquared(length_scale=1.0, periodicity=1.0,
                       periodicity_bounds="fixed"))

# medium term irregularities
k3 = 0.5**2 * RationalQuadratic(length_scale=1.0, alpha=1.0)

# noise terms
k4 = (0.1**2 
      * RBF(length_scale=0.1) 
      + WhiteKernel(noise_level=0.1**2, noise_level_bounds=(1e-3, np.inf)))

kernel = k1 + k2 + k3 + k4
kernel
```

```python
%%time

gp = GaussianProcessRegressor(kernel=kernel, alpha=0,
                              normalize_y=True)
gp.fit(X, y)
```

```python
## 
X_ = np.linspace(X.max(), X.max() + 30, 1000)[:, np.newaxis]
y_pred, y_std = gp.predict(X_, return_std=True)

## Setting colors
c_observed = rgb_2_plt_tuple((255,105,180)) # green -> 83,229,57, purple -> 194,53,150
c_forecast = 'cyan'
c_background = rgb_2_plt_tuple((38,31,66))

c_text = 'white'
mpl.rcParams['text.color'] = c_text
mpl.rcParams['axes.labelcolor'] = c_text
mpl.rcParams['xtick.color'] = c_text
mpl.rcParams['ytick.color'] = c_text
mpl.rc('axes', edgecolor=c_text)

## Plotting
fig, ax = plt.subplots(dpi=150, facecolor=c_background)

ax.plot(X, y, c=c_observed, linewidth=0.75)
ax.plot(X_, y_pred, color=c_forecast, linewidth=0.75)
plt.fill_between(X_[:, 0], y_pred - y_std, y_pred + y_std, 
                 color=c_forecast, alpha=0.5)

hide_spines(ax)
ax.set_facecolor(c_background)
ax.set_xlim(X.min(), X_.max())
ax.set_ylabel('CO$_2$ in ppm')
ax.set_title('Atmospheric CO$_2$ concentration at Mauna Loa', y=0.92)
fig.tight_layout()
```
