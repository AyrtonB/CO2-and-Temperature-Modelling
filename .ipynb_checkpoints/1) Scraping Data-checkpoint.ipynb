### Helper Functions

```python
def hide_spines(ax, positions=['top', 'right']):
    """
    Pass a matplotlib axis and list of positions with spines to be removed
    
    args:
        ax:          Matplotlib axis object
        positions:   Python list e.g. ['top', 'bottom']
    """
    assert isinstance(positions, list), 'Position must be passed as a list '
    
    for position in positions:
        ax.spines[position].set_visible(False)
        
def drop_ticks(ax):
    ax.set_xticks([])
    ax.set_yticks([])

def try_accept(exception=np.nan):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            try:
                return func(*args, **kwargs)
            except:
                return exception
        return wrapper
    return decorator
```

### Forecasted Temperatures

<b>Processing Article</b>

We start by downloading and saving the raw HTML from a Carbon Brief article which includes data on recent IPCC temperature models

```python
url = 'https://www.carbonbrief.org/analysis-how-well-have-climate-models-projected-global-warming'

headers = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'
}

r = requests.get(url, headers=headers)

with open('data/html/carbon_brief_article.html', 'w', encoding='utf-8') as file:
    file.write(r.text)
```

<br>

Next we'll inspect the first 1000 elements of the HTML string to check that the file was downloaded OK

```python
with open('data/html/carbon_brief_article.html', 'r', encoding='utf-8') as file:
    CB_article_HTLM = file.read()
    
CB_article_HTLM[:1000]
```

<br>

Knowing that Carbon Brief chart data is often stored via Amazon AWS we can search for any urls heading that way; sure enough we find links to the HTML for each graph.

```python
str_start = 'https://s3-eu-west-1.amazonaws.com'
str_end = '.html'

highchart_url_mids = re.findall(f'{str_start}(.*?){str_end}', CB_article_HTLM)
highchart_urls = [f'{str_start}{url_mid}{str_end}' for url_mid in highchart_url_mids]

highchart_urls
```

<br>

In this instance the highcharts that these links contain refer to data in a Google spreadsheet, we can access these via the sheet API keys. Here we'll extract the keys from the JS scripts in the HTML.

```python
def url_2_API_key(url):
    ## Retrieving HTML with API key
    r = requests.get(url)
    r_text = r.text

    ## Identifying the API key location
    str_start = 'googleSpreadsheetKey: '
    str_end = ','

    str_start_idx = r_text.find(str_start)
    str_end_idx = r_text[str_start_idx:].find(str_end)

    ## Extracting the API key
    API_key = r_text[str_start_idx+len(str_start)+1:str_start_idx+str_end_idx-1]

    return API_key

url = highchart_urls[0]
API_key = url_2_API_key(url)

API_key
```

<br>

We'll repeat this for each of the chart urls and then remove any duplicate keys (looks like the chart data was doubled up)

```python
%%time

API_keys = [url_2_API_key(highchart_url) for highchart_url in highchart_urls]
API_keys = list(set(API_keys))

API_keys
```

<br>

<b>Retrieving Data</b>

Using the API keys we can now retrieve and parse the data within the charts

```python
sheet_key_2_url = lambda key: f'https://docs.google.com/spreadsheets/d/{key}/pubhtml'

def API_key_2_df(API_key):
    sheet_url = sheet_key_2_url(API_key)

    r_text = requests.get(sheet_url).text
    df_sheet = pd.read_html(r_text, header=1, index_col=0)[0].set_index('year')

    return df_sheet

API_key = API_keys[0]
df_CB_sheet = API_key_2_df(API_key)

df_CB_sheet.head()
```

```python
df_CB = pd.DataFrame()

for API_key in API_keys:
    df_CB_sheet = API_key_2_df(API_key)
    df_CB[df_CB_sheet.columns] = df_CB_sheet

df_CB = df_CB.rename(columns={'Scenaro B':'Scenario B'}) 

df_CB.head()
```

```python
## Observed temperatures
observed_temps = df_CB.columns[:5].to_list()

## Literature forecast temperatures
literature_scenarios = ['Broecker 1975']

scenarios = ['1', '2a', 'A', 'B', 'C']
literature_scenarios += [f'Scenario {scenario}' for scenario in scenarios]

## IPCC forecast temperatures
IPCC_assessments = ['First Assessment', 'Second Assessment', 'Third Assessment', 'Fourth Assessment', 'AR5']
scenarios = ['Low', 'Mean', 'High']

IPCC_assessment_scenarios = [f'{IPCC_assessment} {scenario}' for IPCC_assessment, scenario in itertools.product(IPCC_assessments, scenarios)] 
IPCC_assessment_scenarios += ['AR5 Mean Blended']

## Check accounted for columns
assert len(set(df_CB.columns) - set(observed_temps+literature_scenarios+IPCC_assessment_scenarios)) == 0
```

```python
df_CB.to_csv('data/temperatures/carbon_brief.csv')
```

### Forecasted Temperatures

<b>NASA Data Processing</b>

https://data.giss.nasa.gov/gistemp/

```python
NASA_url = 'https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.csv'
df_NASA = pd.read_csv(NASA_url, skiprows=1, index_col='Year')

df_NASA.head()
```

```python
month_cols = df_NASA.columns[:12]

df_NASA = (df_NASA
           [month_cols]
           .unstack()
           .reset_index()
           .rename(columns={'level_0':'month', 'Year':'year', 0:'temp_dev'})
          )

df_NASA.head()
```

```python
NASA_dts = pd.to_datetime(df_NASA['year'].astype(str) + df_NASA['month'] + '15', format='%Y%b%d')

s_NASA = (pd.Series(df_NASA['temp_dev'].values, index=NASA_dts)
          .sort_index()
          [:-1]
          .astype(float)
         )

s_NASA
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
    
ax.plot(s_NASA, color='k', linewidth=0.75)
    
ax.set_ylabel('Deviation from 1951-1980 Baseline')
ax.set_xlim(s_NASA.index.min(), s_NASA.index.max())
hide_spines(ax)
```

### MET Data Processing

https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/download.html#regional_series

Details of the file format can be found <a href="https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/series_format.html">here</a>

```python
MET_url = 'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.monthly_ns_avg.txt'

r = requests.get(MET_url)
r_text = r.text.replace('   ', ', ')

print('\n'.join(r_text.split('\n')[:5]))
```

```python
## Extracting data, parsing to dataframe and filtering for relevant columns
MET_data = [row.split(', ') for row in r_text.split('\n')[:-1]]
df_MET = pd.DataFrame(MET_data)
df_MET = df_MET[[0, 1, 10, 11]]

## Renaming columns
columns = ['date', 'median', '95pct_lower', '95pct_upper']
df_MET.columns = columns

## Handling dates
df_MET['date'] = pd.to_datetime(df_MET['date'] + '/15', format='%Y/%m/%d')
df_MET = df_MET.set_index('date')
df_MET = df_MET.astype(float)

df_MET.head()
```

```python
dt_plot = df_MET

## Plotting
fig, ax = plt.subplots(dpi=150)

ax.plot(dt_plot['median'], color='k', linewidth=0.75)
ax.fill_between(dt_plot.index, dt_plot['95pct_lower'], dt_plot['95pct_upper'], color='k', alpha=0.25)

ax.set_xlim(dt_plot.index.min(), dt_plot.index.max())
ax.set_ylabel('Deviation from 1961-1990 Baseline')
hide_spines(ax)
```

### Cowtan & Way

http://www-users.york.ac.uk/~kdc3/papers/coverage2013/series.html

The temperature file contains 5 columns: date, temperature, total uncertainty (1$\sigma$), coverage uncertainty (1$\sigma$), and ensemble uncertainty (1$\sigma$).

```python
CandW_url = 'http://www-users.york.ac.uk/~kdc3/papers/coverage2013/had4_krig_v2_0_0.txt'

r = requests.get(CandW_url)
r_text = r.text.replace('  ', '   ').replace('   ', ', ')

print('\n'.join(r_text.split('\n')[:5]))
```

```python
## Extracting data, parsing to dataframe and filtering for relevant columns
CandW_data = [row.split(', ') for row in r_text.split('\n')[:-1]]
df_CandW = pd.DataFrame(CandW_data)

## Renaming columns
columns = ['date', 'temperature', 'total_uncertainty', 'coverage_uncertainty', 'ensemble_uncertainty']
df_CandW.columns = columns

df_CandW.head()
```

```python
year_n_DoY_2_datetime = lambda year, DoY: datetime.date(year, 1, 1) + datetime.timedelta(DoY - 1)

s_year = df_CandW['date'].str[:4].replace('', np.nan).astype(float).astype(int)
s_DoY = (df_CandW['date'].str[4:].replace('', np.nan).astype(float)*365).round().astype(int)

dt_idx = pd.to_datetime([year_n_DoY_2_datetime(year, DoY) for year, DoY in list(zip(s_year, s_DoY))])
df_CandW['date'] = dt_idx
df_CandW = df_CandW.set_index('date')
df_CandW = df_CandW.astype(float)

df_CandW.head()
```

```python
dt_plot = df_CandW#['1975':]

## Plotting
fig, ax = plt.subplots(dpi=150)

ax.plot(dt_plot['temperature'], color='k', linewidth=0.75)
ax.fill_between(dt_plot.index, dt_plot['temperature']+dt_plot['total_uncertainty']*2, dt_plot['temperature']-dt_plot['total_uncertainty']*2, color='k', alpha=0.25)

ax.set_xlim(dt_plot.index.min(), dt_plot.index.max())
ax.set_ylabel('Deviation from 1961-1990 Baseline')
hide_spines(ax)
```

We'll start by downloading the CO2 data from the ESRL site

```python
CO2_url = 'ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_weekly_mlo.txt'

with urllib.request.urlopen(CO2_url) as r:
    CO2_data = r.read()

with open('data/CO2/co2_weekly_mlo.txt', 'w') as file:
    file.write(CO2_data.decode("utf-8") )
```

<br>

The data from Mauna Loa is provided in a text file. The first 49 lines are just information about it's origin and format, the separator used for the data is one or more blank spaces, and NaN values are given as -999.99.

```python
col_names = ['year', 'month', 'day', 'dec_date', 'co2_ppm', 'days', '1yr_ago', '10yr_ago', 'change_from_1800']
df_CO2 = pd.read_csv('data/CO2/co2_weekly_mlo.txt', sep='\s+', skiprows=49, header=None, names=col_names, na_values=-999.99)

df_CO2.head()
```

<br>

Next we'll combine the datetime columns into a single list of datetimes.

Pandas' 'to_datetime()' normally takes something like '2018-12-31' or a list/series in that format, but we can also pass it dictionarys matching each individual datetime component to a series and it will still form the dates.

```python
datetime_dict = {
    'year' : df_CO2.year, 
    'month' : df_CO2.month, 
    'day' : df_CO2.day
}

s_datetimes = pd.to_datetime(datetime_dict)
s_datetimes.head()
```

<br>

We can now create a series where the CO2 concentration is the values and the datetime acts as an index

```python
s_co2 = pd.Series(data=df_CO2['co2_ppm'].values, index=s_datetimes)
s_co2.head()
```

<br>

We need to handle the missing values

```python
s_co2.isnull().sum()
```

```python
s_co2[s_co2.isnull()]
```

```python
s_co2 = s_co2.interpolate()

s_co2.head()
```

```python
s_co2.isnull().sum()
```

```python
s_co2.index.name = 'datetime'
s_co2.name = 'CO2_ppm'

s_co2.to_csv('data/CO2/cleaned_mauna_loa.csv', header=True)
```

```python
fig, ax = plt.subplots(dpi=150) 

s_co2.plot(ax=ax, color='k', linewidth=0.75)

ax.set_title('Rising CO$_{2}$ Levels') # We can use $ symbols around LaTeX formatted stings to do math formatting 
ax.set_ylabel('CO$_{2}$ Concentration (ppm)')
hide_spines(ax)
```
